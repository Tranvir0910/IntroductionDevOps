### **Vanity metrics vs. actionable metrics**

**Reduce Time to Market (Giảm Thời Gian Đưa Sản Phẩm Ra Thị Trường):**

- Đo lường thời gian mà một sản phẩm mới mất từ khi bắt đầu phát triển đến khi nó có thể sẵn sàng để khách hàng sử dụng. Việc giảm thời gian này có thể tăng cơ hội tiếp cận thị trường và đáp ứng nhanh chóng các yêu cầu của khách hàng.

**Increase Overall Availability (Tăng Tổng Khả Dụng):**

- Đo lường khả năng của sản phẩm hoặc dịch vụ để luôn sẵn sàng và hoạt động. Việc nâng cao khả dụng tổng thể giúp đảm bảo rằng khách hàng có thể sử dụng sản phẩm của bạn mọi lúc.

**Reduce the Time to Deploy (Giảm Thời Gian Triển Khai):**

- Đo lường thời gian giữa khi một bản phát hành hoặc tính năng mới sẵn sàng và khi nó được triển khai và sẵn sàng cho người dùng. Việc giảm thời gian triển khai có thể tăng tính linh hoạt và giảm thời gian đáp ứng đối với yêu cầu thị trường.

**Defects Detected Before Production (Lỗi Phát Hiện Trước Sản Xuất):**

- Đo lường tỷ lệ lỗi hoặc khiếm khuyết được phát hiện trong quá trình kiểm thử trước khi sản phẩm hoặc tính năng được triển khai vào môi trường sản xuất. Việc giảm số lượng lỗi trước khi sản xuất giúp đảm bảo tính ổn định và chất lượng của sản phẩm.

**More Efficient Use of Infrastructure (Sử Dụng Hiệu Quả Cơ Sở Hạ Tầng):** 

- Đo lường khả năng của tổ chức để sử dụng hiệu quả các nguồn lực cơ sở hạ tầng như máy chủ, lưu trữ và mạng. Việc tối ưu hóa sử dụng cơ sở hạ tầng giúp giảm chi phí và tăng lợi nhuận.

**Quicker Performance Feedback (Phản Hồi Hiệu Suất Nhanh Chóng):** 

- Đo lường thời gian mà một tổ chức cung cấp thông tin về hiệu suất và trải nghiệm người dùng cho người quản lý sản phẩm. Việc có thông tin nhanh chóng giúp đưa ra quyết định có kiến thức và cải thiện sản phẩm theo thời gian.

**Mean Lead Time (Thời Gian Trung Bình để Dẫn Dắt):** 

- Đo lường thời gian mà một ý tưởng hoặc tính năng mất từ khi yêu cầu được đưa ra đến khi nó được triển khai và sẵn sàng sử dụng.

**Release Frequency (Tần Suất Phát Hành):** 

- Xác định tần suất mà bạn có thể phát hành các phiên bản, bản cập nhật, hoặc tính năng mới. Một tần suất phát hành cao có thể đồng nghĩa với khả năng phản ứng nhanh chóng và linh hoạt.

**Change Failure Rate (Tỷ Lệ Thất Bại của Thay Đổi):** 

- Đo lường tỷ lệ mà các thay đổi, cập nhật, hoặc tính năng mới thất bại khi triển khai. Số liệu này giúp đánh giá tính ổn định và chất lượng của quá trình triển khai.

**Mean Time to Recovery (Thời Gian Trung Bình để Khôi Phục - MTTR):** 

- Đo lường thời gian mà hệ thống mất để khôi phục sau khi xảy ra sự cố hoặc lỗi. MTTR là quan trọng để đảm bảo rằng bạn có thể phục hồi nhanh chóng và giảm thiểu ảnh hưởng đến người dùng và doanh nghiệp.

Các số liệu này cung cấp cái nhìn chi tiết về hiệu suất và chất lượng của quá trình phát triển và triển khai, giúp tổ chức DevOps đưa ra các quyết định thông minh và liên tục cải tiến quy trình làm việc của họ.

### **Comparison of DevOps to Site Reliability Engineering**

#### **Tenets of SRE**

**Chỉ Tuyển Dụng Kỹ Sư Phần Mềm:**

- SRE chủ yếu tập trung vào việc tự động hóa các nhiệm vụ thông qua mã hóa cơ sở hạ tầng.

- Nguyên tắc này nhấn mạnh việc chỉ tuyển dụng những người có kinh nghiệm trong lập trình để có thể tự động hóa các nhiệm vụ lặp đi lặp lại thông qua Mã hóa Cơ sở Hạ tầng.

**Kỹ Sư Độ Tin Cậy Trang Web Giảm Công Việc Đơn Điệu Bằng Tự Động Hóa:**

- Mục tiêu chính của SRE là giảm công việc đơn điệu thông qua việc tự động hóa.

- Đề xuất rằng khoảng 50% thời gian của SRE nên được dành cho việc giảm công việc đơn điệu thông qua tự động hóa.

**Nhóm SRE Tách Biệt Từ Nhóm Phát Triển:**

- Trong SRE, có sự tách biệt giữa nhóm phát triển và nhóm vận hành.

- Sự tách biệt này khác biệt lớn giữa SRE và DevOps, với DevOps nhấn mạnh sự tích hợp giữa nhóm phát triển và nhóm vận hành.

**Tính Ổn Định Được Kiểm Soát Qua Ngân Sách Lỗi:**

- Sự ổn định trong SRE được duy trì thông qua khái niệm của "ngân sách lỗi."

- Nhà phát triển được phép triển khai ứng dụng vào sản xuất miễn là chúng không gây ra quá nhiều sự cố sản xuất, và có một ngân sách lỗi để giới hạn mức độ sự cố được phép.

**Nhà Phát Triển Luân Phiên Qua Vận Hành:**

- Một phần của mô hình SRE là để nhà phát triển hiểu về công việc hàng ngày của nhóm vận hành, họ sẽ dành khoảng 5% thời gian của họ để luân phiên qua vận hành.

- Nếu nhà phát triển gây quá nhiều sự cố hoặc công việc đơn điệu, họ có thể được chuyển sang vận hành để giúp duy trì cân bằng.

#### **Team differences (Sự khác biệt)**

**Nhóm SRE Duy Trì Các Ô Cô Lập Riêng Biệt:**

- Trong mô hình SRE, có sự duy trì của các ô cô lập giữa nhóm phát triển và nhóm vận hành.

- Tuy nhiên, có một bể nhân sự chung, có nghĩa là nếu cần thêm một kỹ sư độ tin cậy trang web, một nhà phát triển có thể được chuyển sang vị trí đó và ngược lại.

- Mô hình này nhấn mạnh việc duy trì sự chuyên nghiệp trong cả hai lĩnh vực, nhưng vẫn có sự chia sẻ nguồn nhân lực.

**DevOps Phá Vỡ Các Ô Cô Lập Thành Một Đội:**

- DevOps chủ yếu tập trung vào sự tích hợp và làm việc chung giữa nhóm phát triển và nhóm vận hành.

- Mô hình này phá vỡ các ô cô lập và tạo ra một đội duy nhất với một mục tiêu kinh doanh chung - triển khai phần mềm vào sản xuất nhanh chóng và an toàn.

- Mỗi thành viên trong đội có trách nhiệm chia sẻ trách nhiệm cho mã nguồn và quản lý quy trình triển khai.

Những sự khác biệt này tương ứng với cách mà SRE và DevOps tiếp cận vấn đề của sự tích hợp giữa các quy trình phát triển và vận hành để đạt được mục tiêu chung của việc triển khai phần mềm một cách hiệu quả và ổn định.

#### **Maintaining stability (Duy trì sự ổn định)**


Cách duy trì tính ổn định trong **mô hình Kỹ thuật Tính sẵn sàng Hệ thống (SRE)** và **phương pháp DevOps** có những điểm khác biệt quan trọng:

##### **SRE (Kỹ thuật Tính sẵn sàng Hệ thống):**

**Ngân Sách Lỗi Dựa Trên Mục Tiêu Mức Dịch Vụ:**

- SRE sử dụng khái niệm "ngân sách lỗi" dựa trên các mục tiêu mức dịch vụ (service-level objectives - SLOs).

- Nhà phát triển được phép triển khai ứng dụng vào sản xuất miễn là không gây ra quá nhiều sự cố sản xuất, và có một ngân sách lỗi để giới hạn mức độ sự cố được phép.

- Khi nhà phát triển vượt quá ngân sách lỗi, họ không được phép triển khai vào sản xuất nữa.

##### **DevOps:**

**CD pipelines**

- DevOps duy trì tính ổn định thông qua việc sử dụng Continuous Delivery pipelines, một quy trình tự động hóa từ việc kiểm tra đến triển khai.

- Mỗi lần có thay đổi mã nguồn, nó có thể đi qua một chuỗi các bước kiểm tra tự động trước khi được triển khai.

- Điều này giúp đảm bảo rằng mã nguồn được kiểm tra và kiểm soát một cách tự động trước khi đi vào môi trường sản xuất.

**Mọi Người Chịu Trách Nhiệm Cho Mã Nguồn:**

- DevOps thúc đẩy nguyên tắc "You build it, you run it" (Bạn xây dựng nó, bạn chạy nó).

- Mỗi thành viên trong đội, bao gồm cả nhóm phát triển và nhóm vận hành, chịu trách nhiệm cho mã nguồn và quy trình triển khai của mình.

- Điều này tạo ra một tinh thần tự chủ và chịu trách nhiệm trong việc duy trì và vận hành ứng dụng.

#### **Commonality**

##### **Điểm Chung giữa SRE và DevOps:**

##### **Tạo Sự Hiển Thị Công Việc giữa Phát Triển và Vận Hành:**

- Cả SRE và DevOps đều hướng đến việc làm cho công việc của nhóm phát triển và nhóm vận hành trở nên rõ ràng đối với nhau.

- Trong SRE, nhà phát triển có thể xoay vòng qua nhóm vận hành để hiểu về công việc hàng ngày của họ.

- Trong DevOps, cả hai nhóm là một, chia sẻ mục tiêu kinh doanh chung để triển khai phần mềm vào sản xuất một cách nhanh chóng và an toàn.

##### **Văn Hóa Không Đổ Lỗi:**

- Cả SRE và DevOps đều yêu cầu một văn hóa không đổ lỗi.

- Trong môi trường không đổ lỗi, người ta có thể nói mở cửa và trung thực về các vấn đề và cách để cải thiện.

- Cả hai phương pháp nhấn mạnh rằng lỗi thường là do hệ thống hoặc quy trình, không phải do người.

##### **Mục Tiêu Cuối Cùng Là Triển Khai Phần Mềm Nhanh Chóng và Ổn Định:**

- Cả SRE và DevOps có mục tiêu chung cuối cùng là triển khai phần mềm một cách nhanh chóng và ổn định.

- Sự tích hợp giữa phát triển và vận hành, cùng với việc tự động hóa quy trình, là để đảm bảo rằng sản phẩm có thể được triển khai một cách hiệu quả và không gây ra sự cố lớn trong môi trường sản xuất.


Những điểm chung này cho thấy sự hội nhập giữa SRE và DevOps trong việc tạo ra môi trường làm việc hiệu quả, trong đó cả hai nhóm đều có hiểu biết và tôn trọng công việc của nhau để đạt được mục tiêu cuối cùng của việc triển khai phần mềm một cách nhanh chóng và ổn định.

#### **DevOps + SRE**

**Kỹ thuật Tính sẵn sàng Hệ thống (SRE):**

- SRE chủ yếu tập trung vào duy trì cơ sở hạ tầng, bao gồm việc quản lý và bảo dưỡng các thành phần cơ bản của hệ thống.

- Nhóm SRE cung cấp và duy trì môi trường đám mây, bao gồm các dịch vụ và tài nguyên cần thiết để chạy ứng dụng.

**DevOps:**

- DevOps sử dụng cơ sở hạ tầng do nhóm SRE duy trì để triển khai và duy trì các ứng dụng của họ.

- Trong môi trường đám mây, DevOps bao gồm những người tiêu thụ các dịch vụ đám mây, sử dụng cơ sở hạ tầng được cung cấp bởi nhóm SRE.

**Tương Tác:**

- Sự tương tác giữa SRE và DevOps được mô tả như việc SRE cung cấp và duy trì nền tảng, trong khi DevOps sử dụng nền tảng đó để duy trì ứng dụng.

- DevOps, thông qua việc triển khai ứng dụng của họ, là người "tiêu thụ" cơ sở hạ tầng được cung cấp bởi SRE.

**Quan Hệ Với Đám Mây:**

- SRE bao gồm những người vận hành đám mây, trong khi DevOps bao gồm những người tiêu thụ đám mây.

- Mối quan hệ giữa SRE và đám mây hình thành cơ sở để cung cấp nền tảng và dịch vụ cho DevOps sử dụng.


Tóm lại, **mô hình SRE đảm nhận vai trò chủ động trong việc duy trì cơ sở hạ tầng, trong khi DevOps tận dụng cơ sở hạ tầng này để duy trì và triển khai ứng dụng của họ**. Mối quan hệ giữa SRE và DevOps là quan trọng để đảm bảo môi trường hoạt động hiệu quả và ổn định.
