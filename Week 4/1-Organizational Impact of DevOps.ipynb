### **DevOps is not job title**

- Các tổ chức cần có các nhóm nhỏ, chuyên dụng, chức năng chéo, tự tổ chức để triển khai DevOps thành công.

- Định luật Conway ngụ ý rằng kết quả thiết kế của một công ty là phản ánh trực tiếp của cấu trúc giao tiếp của công ty.

- Thay vì cấu trúc truyền thống được tổ chức xung quanh công nghệ, các nhóm DevOps thành công nên được tổ chức xung quanh lĩnh vực kinh doanh. Mỗi nhóm nên có nhiệm vụ riêng phù hợp với một lĩnh vực kinh doanh.

- DevOps là một tư duy mà toàn bộ tổ chức áp dụng.

- DevOps giải quyết các vấn đề do các nhóm hoạt động độc lập gây ra.

- DevOps là thực hành của các kỹ sư phát triển và vận hành làm việc cùng nhau trong suốt vòng đời phần mềm, tuân thủ các nguyên tắc lean và Agile cho phép họ cung cấp kết quả chất lượng cao.

- Hành động không có hậu quả có thể dẫn đến sự thờ ơ.

- Cho phép các nhóm cảm nhận tác động của hành động của họ tạo điều kiện cho sự đồng cảm, dẫn đến công việc chất lượng cao hơn.

- Mục tiêu tổ chức của DevOps là đạt được tư duy chung và trao quyền cho mọi người để cung cấp giá trị cho khách hàng.
