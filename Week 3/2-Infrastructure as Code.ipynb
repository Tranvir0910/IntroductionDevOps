## **Infrastructure as Code**

- Resource: [Infrastructure as Code](https://viblo.asia/p/infrastructure-as-code-la-gi-tai-sao-lai-chon-terraform-LzD5dwrEljY)

**'Infrastructure as code'** là thực hành mô tả hạ tầng dưới dạng văn bản. một định dạng văn bản có thể thực thi, còn được biết đến là **Code**. 

Hiện nay, việc quản lý các **web applications** khá là phức tạp, ví dụ bạn phải quản lý **load balancer, vài web servers, database servers, ...** Bạn có thể sẽ phải mất hàng giờ để provisioning và quản lý những hệ thống này. 

Infrastructure as Code có thể hiểu là bạn có thể thiết lập/quản lý những stack trước kia của hệ thống thông qua việc định nghĩa chúng trong 1 file script chẳng hạn thay vì tốn thời gian và công sức setup manual từng thứ (eg: ssh rồi cài cắm package, services, lib nọ kia).

#### **Described an executable textual format**

- Một định dạng văn bản có thể thực thi là một cách biểu diễn thông tin dưới dạng văn bản mà có thể được máy tính hiểu và thực thi. Trong ngữ cảnh của **"infrastructure as code"** và **quản lý cấu hình**, định dạng này là một phương tiện để **mô tả cụ thể và chi tiết các cấu hình hạ tầng một cách thực thi được**.

#### **Configure using that description**

- Việc cấu hình sử dụng mô tả này liên quan đến việc áp dụng các thay đổi, cài đặt, và cấu hình hệ thống, thiết bị, phần mềm, và người dùng theo các thông số được đặc tả trong định dạng văn bản. Thay vì thực hiện các thay đổi này thủ công, bạn có thể sử dụng một công cụ tự động hóa để đọc và thực thi định dạng văn bản này, giúp đảm bảo tính chính xác và lặp lại của quá trình cấu hình.

- Công nghệ như **Docker, Vagrant, Terraform, và thậm chí Kubernetes** cũng cho phép bạn mô tả hạ tầng của mình dưới dạng mã nguồn, và mã nguồn này nên được kiểm tra vào hệ thống quản lý phiên bản.

#### **Configuration Management Systems make this possible**

- Hệ thống quản lý cấu hình (Configuration Management Systems) là các công cụ được thiết kế để hỗ trợ việc triển khai, cấu hình, và duy trì các thành phần hạ tầng trong một môi trường máy chủ. Các công cụ như Ansible, Puppet, và Chef là những ví dụ về Configuration Management Systems. Chúng cho phép bạn mô tả cấu hình của hạ tầng dưới dạng mã nguồn thực thi được, sau đó thực hiện các thay đổi và duy trì trạng thái của hạ tầng đó.

#### **Never perform configurations manually**

- Việc không bao giờ thực hiện cấu hình thủ công là quan trọng để đảm bảo sự tái tạo được và tránh lỗi do con người. Thay vào đó, bạn sử dụng các mẫu và kịch bản mà mô tả cách thức cài đặt và cấu hình tự động các yếu tố trong hạ tầng.

#### **Use version control**

- Việc lưu trữ mã nguồn của cấu hình trong hệ thống quản lý phiên bản (version control) như Git là quan trọng để theo dõi và quản lý lịch sử các thay đổi. Điều này giúp đội ngũ phát triển và quản trị hệ thống theo dõi phiên bản nào là phiên bản mới nhất và làm thế nào hạ tầng nên trông như thế nào ở từng điểm thời gian.

## **Ephemeral immutable infrastructure**

**Ephemeral immutable infrastructure (Cơ sở hạ tầng tạm thời không thể thay đổi.)** là một cách tiếp cận trong quản lý hạ tầng và triển khai ứng dụng.

**Ephemeral Immutable Infrastructure (Hạ tầng Tạm Thời và Bất Biến):**

- Hạ tầng tạm thời ám chỉ rằng các thành phần của hệ thống chỉ tồn tại trong khoảng thời gian cần thiết và sau đó được loại bỏ khi không còn sử dụng.

- Bất biến ám chỉ rằng một khi hạ tầng được triển khai, nó không thể bị thay đổi. Thay vào đó, nếu cần thực hiện thay đổi, bạn xây dựng một phiên 
bản mới thay thế phiên bản cũ.

**Server Drift as a Major Source of Failure (Sự Lệch Lạc của Máy Chủ là Nguồn Gốc Chính của Sự Cố):**

- Server drift là hiện tượng khi các máy chủ trên hạ tầng không giữ được trạng thái cấu hình ban đầu và lệch khỏi cấu hình mong muốn.

- Việc này có thể gây ra sự cố không dự đoán được do sự khác biệt giữa các máy chủ.

**Servers are Cattle, Not Pets (Máy Chủ như Gia Súc, Không Phải Là Thú Cưng):**

- So sánh này nhấn mạnh cách bạn xử lý máy chủ. Thay vì chăm sóc và tu sửa từng máy chủ cá nhân (như thú cưng), máy chủ được coi như là một nhóm (gia súc) và có thể được thay thế một cách linh hoạt.

**Infrastructure is Transient (Hạ Tầng Là Tạm Thời):**

- Điều này nói về tính chất tạm thời của hạ tầng. Nó chỉ tồn tại khi cần thiết và có thể được xây dựng và hủy bỏ theo yêu cầu.

**Build Through Parallel Infrastructure (Xây Dựng Qua Cơ Sở Hạ Tầng Song Song):**

- Việc xây dựng qua cơ sở hạ tầng song song ám chỉ việc triển khai và kiểm thử các phiên bản mới của hạ tầng song song với hạ tầng đang chạy, giúp đảm bảo tính ổn định và chất lượng của hệ thống khi triển khai thay đổi.

## **Immutable delivery via containers**

**Immutable delivery via containers** là một phương pháp triển khai mà sử dụng khái niệm về sự bất biến và sử dụng container.

**Applications are Packaged in Containers (Ứng Dụng Được Đóng Gói trong Container):**

- Ứng dụng và tất cả các phụ thuộc của nó (dependencies) được đóng gói và cô lập trong một container.

**Same Container that Runs in Production Can be Run Locally (Container chạy trong môi trường sản xuất có thể chạy ở địa phương):**

- Container cung cấp môi trường cô lập, giúp đảm bảo rằng cùng một container có thể chạy trên môi trường phát triển hay máy tính cá nhân của nhà phát triển mà không có sự thay đổi.

**Dependencies are Contained (Các Phụ Thuộc Được Bao Gồm):**

- Tất cả các thành phần cần thiết để chạy ứng dụng (dependencies) được đóng gói cùng với ứng dụng trong container. Điều này giúp tránh được vấn đề về sự không đồng nhất trong môi trường.

**No Variance Limits Side-effects (Không Sự Biến Đổi Giới Hạn Tác Động Phụ):**

- Container giúp giữ cho môi trường thực thi không thay đổi qua thời gian, giảm nguy cơ lỗi do sự biến đổi không kiểm soát.

**Rolling Updates with Immediate Roll-back (Cập Nhật Cuộn với Khả Năng Quay Lại Ngay Lập Tức):**

- Việc triển khai các cập nhật theo kiểu cuộn giúp giảm thiểu thời gian mà hệ thống không hoạt động đầy đủ. Nếu có sự cố xảy ra, quay lại phiên bản trước đó của ứng dụng là một quyết định ngay lập tức.

Thông qua việc sử dụng containers, có thể thấy rõ ưu điểm của việc triển khai và quản lý hệ thống một cách đồng nhất, linh hoạt và an toàn. Các nguyên tắc này hỗ trợ mô hình "infrastructure as code" và thực hành DevOps, nơi sự tự động hóa và khả năng lăp trình cho cả hạ tầng và ứng dụng đóng một vai trò quan trọng.

## **Immutable way of working**

**Immutable way of working** là một phương pháp làm việc trong quản lý hạ tầng và triển khai ứng dụng mà đặc trưng bởi tính bất biến và việc thực hiện thay đổi không trực tiếp trên container đang chạy.

**You Never Make Changes to a Running Container (Không Bao Giờ Thực Hiện Thay Đổi Trực Tiếp trên Container Đang Chạy):**

- Container đang chạy được coi là không thể thay đổi. Mọi thay đổi cần phải được áp dụng thông qua quá trình xây dựng lại container từ image mới.

**You Make Changes to the Image (Thực Hiện Thay Đổi trên Hình Ảnh):**

- Mọi sự thay đổi, cập nhật, hay vá lỗi đều được thực hiện trên image, không phải container đang chạy. Image là một bản mẫu hoàn chỉnh của môi trường chạy của ứng dụng.

**Then Redeploy a New Container (Sau Đó Triển Khai Một Container Mới):**

- Khi có sự thay đổi trong image, bạn tạo ra một container mới từ image đã được cập nhật và triển khai nó. Điều này đảm bảo rằng mọi phiên bản mới của ứng dụng được triển khai dưới dạng một container mới.

**Keep Images Up to Date (Duy Trì Cập Nhật cho Images):**

- Việc duy trì cập nhật cho các image là quan trọng để đảm bảo rằng mọi container mới được tạo ra từ image mới nhất, chứa tất cả các cập nhật và sửa lỗi mới nhất.

Phương pháp này giúp đảm bảo tính nhất quán giữa môi trường phát triển và môi trường sản xuất, giảm thiểu sự lệch lạc giữa các phiên bản và giảm nguy cơ lỗi do sự biến đổi không kiểm soát. Nó cũng hỗ trợ trong việc triển khai các cập nhật một cách linh hoạt và an toàn.
