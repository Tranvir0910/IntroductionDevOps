### **Required DevOps Behaviors**

**S Hp Tc v S Hu Chung (Shared ownership and high collaboration):**

- Tng cng tinh thn ng i v s hp tc gia cc b phn  m bo mi ngi cm thy c trch nhim v quyn s hu i vi mc tiu chung.

**Qun L Ri Ro bng Cch Chp Nhn Thay i (Risk management by embracing change):**

- Qun l ri ro khng phi qua vic hn ch thay i, m l thng qua vic chp nhn v qun l nhng thay i nh v lin tc, tng cng kh nng ng ph vi s thay i.

**H Tng Tm Thi di Dng M Ngun (Ephemeral infrastructure as code):**

- To v qun l h tng di dng m ngun, tm thi v c th ti to, gip m bo tnh lp li v d qun l ca qu trnh trin khai.

**T ng Ha Dch V (Automated self-service):**

- Thc y vic t ng ha cc quy trnh v dch v  gim bt cng vic th cng, tng cng kh nng t phc v v tng tc qu trnh pht trin.

**Vng Lp Phn Hi v p ng Da Trn D Liu (Feedback loops and data-driven responses):**

- S dng vng lp phn hi nhanh chng, da trn d liu t sn xut  m bo kh nng p ng nhanh chng v hiu qu khi x l s c v thc hin ci tin.
