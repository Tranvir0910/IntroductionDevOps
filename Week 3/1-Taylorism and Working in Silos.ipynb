### **Required DevOps Behaviors**

**Sự Hợp Tác và Sở Hữu Chung (Shared ownership and high collaboration):**

- Tăng cường tinh thần đồng đội và sự hợp tác giữa các bộ phận để đảm bảo mọi người cảm thấy có trách nhiệm và quyền sở hữu đối với mục tiêu chung.

**Quản Lý Rủi Ro bằng Cách Chấp Nhận Thay Đổi (Risk management by embracing change):**

- Quản lý rủi ro không phải qua việc hạn chế thay đổi, mà là thông qua việc chấp nhận và quản lý những thay đổi nhỏ và liên tục, tăng cường khả năng ứng phó với sự thay đổi.

**Hạ Tầng Tạm Thời dưới Dạng Mã Nguồn (Ephemeral infrastructure as code):**

- Tạo và quản lý hạ tầng dưới dạng mã nguồn, tạm thời và có thể tái tạo, giúp đảm bảo tính lặp lại và dễ quản lý của quá trình triển khai.

**Tự Động Hóa Dịch Vụ (Automated self-service):**

- Thúc đẩy việc tự động hóa các quy trình và dịch vụ để giảm bớt công việc thủ công, tăng cường khả năng tự phục vụ và tăng tốc quá trình phát triển.

**Vòng Lặp Phản Hồi và Đáp Ứng Dựa Trên Dữ Liệu (Feedback loops and data-driven responses):**

- Sử dụng vòng lặp phản hồi nhanh chóng, dựa trên dữ liệu từ sản xuất để đảm bảo khả năng đáp ứng nhanh chóng và hiệu quả khi xử lý sự cố và thực hiện cải tiến.
