## **Continuous Integration (Tich hợp liên tục)**

[What is CI/CD](https://www.youtube.com/watch?v=PkNg9gUhiYQ)

- **Continuous Integration (CI)** là quá trình liên tục xây dựng, kiểm thử, và tích hợp mỗi thay đổi từ các nhà phát triển vào nhánh chính sau khi một bộ kiểm thử đã được vượt qua. Kết quả có thể là mã nguồn có thể triển khai.

- **Continuous Delivery (CD)** là một chuỗi các thực hành nhằm đảm bảo rằng mã nguồn có thể được triển khai nhanh chóng và an toàn vào môi trường sản xuất, bằng cách triển khai mỗi thay đổi vào một môi trường giống với môi trường sản xuất.

**Developers Integrate Code Often (Nhà Phát Triển Tích Hợp Mã Thường Xuyên) :**

- Nhà phát triển tích hợp mã nguồn thường xuyên vào kho lưu trữ chung.

**Developers Work in Short-lived Feature Branches (Nhà Phát Triển Làm Việc Trong Nhánh Chức Năng Có Thời Gian Sống Ngắn) :**

- Nhà phát triển làm việc trong các nhánh chức năng có thời gian sống ngắn.

**Each Check-in is Verified by an Automated Build (Mỗi Lần Check-in Đều Được Xác Nhận Bằng Một Quá Trình Xây Dựng Tự Động) :**

- Mỗi lần nhà phát triển thực hiện check-in, quá trình xây dựng tự động được kích hoạt và kiểm tra tích hợp.

## **Ưu Điểm của Continuous Integration**

- Làm việc với các "**small batches**" giúp thực hiện Continuous Integration bằng cách giảm số lượng thay đổi xung đột do thời gian giữa các lần tích hợp ít hơn.

- Sử dụng "**pull requests**" để gửi thay đổi, tăng cường chất lượng mã nguồn qua việc đánh giá mã giảm nguy cơ lỗi.

- Hệ thống tự động xây dựng và kiểm thử mỗi "**pull request**" để đảm bảo rằng thay đổi là hợp lệ và không gây lỗi.

- **Continuous Integration** giảm rủi ro tích hợp mã vì bạn tích hợp các thay đổi nhỏ.

## **Lợi Ích của Continuous Integration**

- Tăng cường nhanh chóng và di chuyển nhanh hơn nhờ vào việc tự động hóa các bước kiểm thử.

- Giảm rủi ro tích hợp mã bằng cách tích hợp các thay đổi nhỏ.

- "**Pull requests**" giúp tăng chất lượng mã thông qua đánh giá của đồng đội.

- Đảm bảo rằng nhánh chính luôn có thể triển khai.

Nhớ rằng Continuous Integration và Continuous Delivery là hai thực hành riêng biệt, và Continuous Integration đặt nền tảng cho Continuous Delivery.

## **Continuous Delivery (Phân Phối Liên Tục)**

**Continuous Integration (CI) là quá trình tích hợp liên tục:**

- CI là kỹ thuật phát triển phần mềm trong đó bạn xây dựng phần mềm sao cho nó có thể được triển khai vào môi trường sản xuất bất kỳ lúc nào.

- Để thực hiện điều này, nhánh chính (master branch) phải luôn có thể triển khai được.

**CI đòi hỏi có quá trình xây dựng và kiểm thử liên tục:**

- Điều này có nghĩa là mỗi thay đổi đều cần được xây dựng và kiểm thử, đặc trưng của Continuous Integration.

**CI/CD Pipeline là một chuỗi công cụ:**

- Có một chuỗi các kiểm tra tự động như kiểm thử đơn vị, kiểm thử tích hợp, kiểm thử chất lượng, quét lỗ hổng và kiểm thử bảo mật.
Được tổ chức dưới dạng một pipeline, nơi đầu ra của mỗi công cụ là đầu vào của công cụ tiếp theo.

## **CI/CD pipeline**

**Kho Lưu Trữ Mã (Code Repository):**

- Nơi quản lý và lưu trữ mã nguồn của dự án.

**Máy Chủ Xây Dựng (Build Server):**

- Cung cấp khả năng xây dựng ứng dụng từ mã nguồn.

**Máy Chủ Tích Hợp/Orchestrator (Integration Server/Orchestrator):**

- Tự động hóa quy trình xây dựng và chạy các kiểm thử chất lượng.

**Kho Lưu Trữ Đối Tượng (Artifact Repository):**

- Lưu trữ các đối tượng như file nhị phân, hình ảnh Docker, hoặc các thành phần cần thiết khác đã được kiểm thử và chứng minh hoạt động.

**Công Cụ Cấu Hình Tự Động và Triển Khai (Automatic Configuration and Deployment Tool):**

- Tự động hóa việc triển khai và cấu hình môi trường triển khai.

**Ghi Chú:**

- Máy chủ xây dựng (Build Server) và máy chủ tích hợp/orchestrator (Integration Server/Orchestrator) thường được tích hợp trong nhiều dịch vụ CI/CD hiện đại.
- Các công cụ như **Travis CI, GitHub Actions, Jenkins, IBM Cloud Continuous Delivery** cung cấp các tính năng này.

## **Five key principles**

**Xây Dựng Chất Lượng Ngay Từ Đầu (Build Quality In):**

- Tự động hóa kiểm thử và kiểm tra chất lượng đối với mỗi thay đổi mã nguồn để đảm bảo chất lượng ngay từ giai đoạn sớm nhất.

**Làm Việc Theo Các Lô Nhỏ (Work in Small Batches):**

- Thực hiện các thay đổi mã nguồn trong các đợt nhỏ để giảm thiểu rủi ro và dễ dàng kiểm thử.

**Máy Tính Thực Hiện Công Việc Lặp (Computers Perform Repetitive Tasks, People Solve Problems):**

- Tự động hóa các công việc lặp để máy tính thực hiện, giúp con người tập trung vào giải quyết vấn đề và công việc sáng tạo.

**Khám Phá Liên Tục (Relentlessly Pursue Continuous Improvement):**

- Không ngừng tìm kiếm cách cải thiện, sử dụng đo lường hành động và thực hiện các biện pháp cải tiến khi cần.

**Mọi Người Chịu Trách Nhiệm (Everyone Is Responsible):**

- Mọi thành viên trong đội ngũ chịu trách nhiệm đối với chất lượng và sự thành công của quá trình phát triển và triển khai. Mọi người đều đóng góp vào việc giữ cho quá trình diễn ra mượt mà và hiệu quả.

**Ghi Chú:**

Các nguyên tắc này là cơ sở của **Continuous Delivery** và hỗ trợ quy trình CI/CD.

## **CI/CD + Continuous Deployment**

**CI**

- **Kho Lưu Trữ Mã & Quản Lý Phiên Bản (Code Repository & Version Control):**

    - Nhà phát triển đẩy mã nguồn và tự động kích hoạt quá trình xây dựng trên máy chủ.

- **Tích Hợp & Tự Động Hóa Xây Dựng (Build Integration & Automation):**

    + Máy chủ CI bắt đầu quá trình xây dựng và chạy kiểm thử tự động.

**CD**

- **Tự Động Hóa Phát Hành (Release Automation):**

    - Lưu trữ kết quả của bất kỳ thành phẩm nào từ quá trình xây dựng.

- **Tự Động Hóa Giao Hàng (Delivery Automation):**

    - Triển khai các đối tượng nhị phân đến bất kỳ môi trường nào được chỉ định.

**Continuous Deployment**

- **Tự Động Hóa Sản Xuất (Production Automation):**

    - Thúc đẩy việc triển khai lên môi trường sản xuất (on-premises, public cloud, hybrid cloud).
